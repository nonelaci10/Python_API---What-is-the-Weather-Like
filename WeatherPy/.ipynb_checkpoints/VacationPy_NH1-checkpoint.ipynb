```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

```python
#Store Part I results into DataFrame
##Load the csv exported in Part I to a DataFrame
```

```python
weather_df1 = pd.read_csv("data_cities(1).csv")
weather_df1.dropna()
weather_df1
```

```python
 # Store latitude and longitude in locations
locations = weather_df1[["Lat", "Lng"]]
humidity = weather_df1["Humidity"]
```

```python
##Humidity Heatmap
#Configure gmaps.
#Use the Lat and Lng as locations and Humidity as the weight.
#Add Heatmap layer to map.
```

```python
 # Plot Heatmap
fig = gmaps.figure()
# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=1)
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig 
```

```python
##Create new DataFrame fitting weather criteria
#Narrow down the cities to fit weather conditions.
#Drop any rows with null values.
```

```python
weather_crit = weather_df1[
    (weather_df1['Max Temp'] < 25)
    & (weather_df1['Max Temp'] > 15)
    & (weather_df1['Wind Speed'] < 10)
    & (weather_df1['Cloudiness'] == 0 )
    ]
del weather_crit['Unnamed: 0']
weather_crit
```

```python
hotel_weather_df = weather_crit.reset_index()
del hotel_weather_df['index']
hotel_weather_df 
```

```python
##Hotel Map
#Store into variable named hotel_df.
#Add a "Hotel Name" column to the DataFrame.
#Set parameters to search for hotels with 5000 meters.
#Hit the Google Places API for each city's coordinates.
#Store the first Hotel result into the DataFrame.
#Plot markers on top of the heatmap.

hotel_column = len(weather_crit) * ['']

index = 0

for hotel in hotel_column:
    weather_crit.loc[index,8] = hotel
    index = index + 1
    
#weather_crit['Hotel Name'] = hotel_column
weather_crit.head()
```

```python
hotel_weather_df['Hotel Name'] = ""
hotel_weather_df
```

```python
# Hotel API Call

#print(str(weather_crit["Lat"]) + "," + str(weather_crit["Lng"]))
#target_coord = str(weather_crit["Lat"]) + "," + str(weather_crit["Lng"])
#latt = str(hotel_weather_df.iloc[2]) 
#lngt = str(hotel_weather_df.iloc[8])
#target_coord = "['latt'], ['lngt']"

hotel_name = []
hotel_vicinity = []

#target_search = "hotel"
#target_radius = 5000
#target_type = "establishment"

params = {
    "radius" : 5000,
    "keyword" : 'hotel',
    "type" : ['establishment', 'lodging'],
    "key" : g_key
}

for index, row in hotel_weather_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    
    params['location'] = f"{lat},{lng}"
    
    hotel_base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

    response = requests.get(hotel_base_url, params=params)

    places_data = response.json()
    #print(json.dumps(places_data, indent=4, sort_keys=True))
    
    try:
        hotel_weather_df.loc[index, "Hotel name"] = places_data["results"][0]["name"]
        #hotel_vicinity.append(places_data['results'][0]['vinicity'])
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    #except:
        #print("no data found...moving on to next")
        #hotel_name.append(None)
        #hotel_vicinity.append(None)

hotel_weather_df
```

```python
hotel_weather_df = hotel_weather_df.dropna(axis = 0)
hotel_weather_df
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_weather_df.iterrows()]
locations1 = hotel_weather_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map

locations = hotel_weather_df[['Lat', 'Lng']]
humidity = hotel_weather_df["Humidity"]

fig1 = gmaps.figure(center=(0,0), zoom_level = 2)

# Heat layer
heat_layer1 = gmaps.heatmap_layer(locations, weights=humidity, opacity=0.5,point_radius=15)

fig1.add_layer(heat_layer1)
marker_layer = gmaps.marker_layer(locations1,info_box_content=hotel_info)

fig1.add_layer(marker_layer)
fig1
# Display Map
```
